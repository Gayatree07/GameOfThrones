```python
#print columns with missing values as a list
def getColumnsWithMissingValuesList(df):
    return[col for col in df.columns if df[col].isnull().any()]
getColumnsWithMissingValuesList(df)
```

```python
#display columns with object data type (categorical columns) as a list
def getObjectColumnsList(df):
    return[col_name for col_name in df.columns if df[col_name].dtype == 'object']
cols = getObjectColumnsList(df)
cols
```

```python
#display columns that contains numeric values as a list
def getNumericColumnsList(df):
    return [col_name for col_name in df.columns if df[col_name].dtype == ['int64','float64']]
num_cols = getNumericColumnsList(df)
num_cols

def getNumericColumnsList(df):
    return [cname for cname in df.columns if df[cname].dtype in ['int64', 'float64']]
```
